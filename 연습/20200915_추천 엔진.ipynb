## 무비렌즈 데이터에 적용한 추천 엔진

```python
# 데이터 가져오기
import pandas as pd
ratings = pd.read_csv('./Data/ml-latest-small/ratings.csv')
movies = pd.read_csv('./Data/ml-latest-small/movies.csv')
```

```python
# 데이터 확인하기 - ratings
ratings.head()
```

```python
# 데이터 확인하기 - movies
movies.head()
```

```python
# 두 데이터에서 movieID가 공통 컬럼 -> 데이터 프레임 통합하기 : merge
ratings = pd.merge(ratings, movies, left_on = 'movieId', right_on = 'movieId').drop(['timestamp', 'genres'], axis = 1)
```

```python
# 데이터 확인
ratings
```

```python
# userId => Index, columns => movieId로 하는 pivot_table 만들기
rp = ratings.pivot_table(index = ['userId'], columns = ['movieId'], values = ['rating'])
```

```python
# 데이터 확인
rp
```

```python
# 결측치를 0으로 변경하자
rp.fillna(0, inplace = True)
```

```python
# 데이터 확인
rp
```

```python
# 계산 속도를 향상 시키기 위해서 DataFrame을 array로 전환 하자
import numpy as np
rp_array = rp.to_numpy()
```

```python
# 데이터 확인
rp_array
```

#### 영화-영화 유사도 행렬

```python
import time
from scipy.spatial.distance import cosine

start_time = time.time()

m, n = rp.shape
movies_array = np.zeros((n,n))
for i in range(n):
    for j in range(n):
        if i != j:
            movies_array[i,j] = 1 - cosine(rp_array[:,i], rp_array[:,j])
        else:
            movies_array[i,j] = 0

            
print(time.time() - start_time)           
```

```python
pd_movies = pd.DataFrame(
    movies_array,
    index = rp.columns,
    columns = rp.columns
)
```

```python
pd_movies.to_csv('./Data/ml-latest-small/pd_movies.csv', sep = ',')
pd_movies = pd.read_csv('./Data/ml-latest-small/pd_movies.csv')
```
