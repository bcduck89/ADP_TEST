```python
# 모듈 가져오기
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### 데이터 - iris

```python
# 데이터 불러오기 - iris
iris = pd.read_csv('./Data/iris.csv')
```

```python
# 1-1. 데이터 확인 - head
iris.head()
```

```python
# 1-2. 데이터 확인 - info()
iris.info()
```

```python
# 1-3 데이터 확인 - 산점도
sns.pairplot(iris, hue = 'class')
```

```python
# 1-4 데이터 확인 - 상관관계
plt.figure(figsize = (10,10))
sns.heatmap(iris.corr(), annot = True, vmin = -1, vmax = 1)
```

```python
# 2-1. 주성분 분석 수행 -  종속변수, 독립변수 데이터 분할
X = iris.drop('class', axis = 1)
y = iris['class']
```

```python
# 2-2. 주성분 분석 수행 - 정규화
from sklearn.preprocessing import StandardScaler
X_scale = pd.DataFrame(StandardScaler().fit_transform(X), columns = X.columns)
```

```python
# 2-3. 주성분 분석 수행 - PCA 모델 생성
from sklearn.decomposition import PCA
n_components = len(X.columns)
pca = PCA(n_components = n_components)
```

```python
# 2-4. 주성분 분석 수행 - PCA fitting
columns = list()
for comp in range(n_components):
    columns.append('PC' + str(comp + 1))
pca_fit = pd.DataFrame(pca.fit_transform(X_scale), columns = columns)
```

```python
# 2-5. 주성분 분석 수행 - 데이터 확인
pca_fit
```

```python
# 2-6 주성분 분석 수행 - 주성분끼리 상관계수 확인
pca_fit.corr()
```

```python
# 2-7. 주성분 분석 수행 - 기여율 시각화

explained_variance_ratio = pca.explained_variance_ratio_

def explained_variance_ratio_plot(explained_variance_ratio):
    x_axis = range(1, len(explained_variance_ratio)+1)
    plt.figure(figsize = (10,5))
    plt.bar(x_axis, explained_variance_ratio,
            align = 'center', label = 'Individual Explained Variance Ratio')
    plt.step(x_axis, np.cumsum(explained_variance_ratio),
             where = 'mid', color='red', label='Cumulative Explained Variance Ratio')
    plt.ylim(0, 1.1)
    plt.xticks(x_axis)
    plt.xlabel('Principal Components')
    plt.ylabel('Explained Variance Ratio')
    plt.grid()
    plt.legend()
    plt.show()
explained_variance_ratio_plot(explained_variance_ratio)
```

```python
# 2-8. 주성분 분석 수행 - 기여율
print('variance : ', pca.explained_variance_)
print('proportion of variance : ', pca.explained_variance_ratio_)
print('Cumulative Proportion : ', pca.explained_variance_ratio_.cumsum())
```

```python
# 2-9. 주성분 분석 수행 - scree plot

xticks = range(1,1+n_components)
xticks_label = ['Comp' + str(x) for x in xticks]

plt.plot(range(1,n_components+1), pca.explained_variance_ratio_, 'ro-')
plt.title('Scree Plot')
plt.xlabel('Principle Components')
plt.ylabel('Proportion of Variance')
plt.xticks(xticks, xticks_label)
plt.ylim([0,1])
plt.grid()
```

```python
# 시각화
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = y.unique()
colors = ['r','g','b']
for target, color in zip(targets, colors):
    index = y == target
    ax.scatter(
        x = pca_fit.loc[index,'PC1'],
        y = pca_fit.loc[index,'PC2'],
        c = color,
        s = 50
    )
ax.grid()
ax.legend(targets)
```
