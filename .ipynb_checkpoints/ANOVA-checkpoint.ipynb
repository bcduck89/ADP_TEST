# 1) One-way ANOVA

```python
# 모듈 가져오기
import pandas as pd
from scipy.stats import f_oneway
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
# 데이터 가져오기 - iris
iris = pd.read_csv('./Data/iris.csv')
```

```python
# 데이터 확인
iris
```

```python
col1 = iris['sepal_length'][iris['class'] == iris['class'].unique()[0]]
col2 = iris['sepal_length'][iris['class'] == iris['class'].unique()[1]]
col3 = iris['sepal_length'][iris['class'] == iris['class'].unique()[2]]
```

```python
# one-way ANOVA - f_oneway
f_oneway(col1,col2,col3)
```

```python
# one-way ANOVA - ols, anova_lm
iris.rename(columns = {'class' : 'what'}, inplace = True)
model = ols("sepal_length ~ C(what)", iris).fit()
anova_lm(model)
```

```python
# 사후분석 - pairwise_tukeyhsd
posthoc = pairwise_tukeyhsd(iris['sepal_length'], iris['what'])
print(posthoc)
```

```python
# 사후분석 - plot_simultaneous()
fig = posthoc.plot_simultaneous()
```

# 2) Two-way ANOVA

```python
# 데이터 가져오기 - titanic
FIFA = pd.read_csv('./Data/data_edu/모의고사 1회/FIFA.csv', encoding = 'cp949')
```

```python
# 데이터 확인
FIFA.info()
```

```python
# 데이터 확인
FIFA
```

```python
formula = 'Wage ~ C(Position) + C(Height) + C(Position):C(Height)'
model = ols(formula, FIFA).fit()
anova_lm(model)
```
