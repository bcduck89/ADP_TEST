```python
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv("./Data/iris.csv")
```

```python
iris.info()
```

```python
iris.columns
```

```python
iris.index
```

```python
iris['class'].unique()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris.drop(['class'], axis = 1), iris['class'], train_size = 0.7, random_state = 42)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_fit = RandomForestClassifier().fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
pd.crosstab(y_train, rf_fit.predict(x_train), rownames = ['Actual'], colnames = ['Predicted'])
```

```python
pd.crosstab(y_test, rf_fit.predict(x_test), rownames = ['Actual'], colnames = ['Predicted'])
```

```python
accuracy_score(y_test, rf_fit.predict(x_test))
```

```python
print(classification_report(y_test, rf_fit.predict(x_test)))
```
