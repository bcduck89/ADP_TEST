```python
# 모듈 가져오기
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 가져오기
iris = pd.read_csv('./Data/iris.csv')
```

```python
# 데이터 확인
iris
```

```python
corr_kendall = iris.corr(method = 'kendall')
corr_pearson = iris.corr(method = 'pearson')
corr_spearman = iris.corr(method = 'spearman')
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(corr_kendall, vmin = -1, vmax = 1, annot = True)
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(corr_pearson, vmin = -1, vmax = 1, annot = True)
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(corr_spearman, vmin = -1, vmax = 1, annot = True)
```

```python
sns.pairplot(iris)
```
